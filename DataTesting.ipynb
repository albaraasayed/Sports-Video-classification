```python
from keras.models import load_model
from collections import deque
import numpy as np
import pickle
import cv2
from google.colab.patches import cv2_imshow
```

```python
model = load_model("/content/drive/MyDrive/output/videoOutput")
lb = pickle.load(open("/content/drive/MyDrive/output/labelsOutput.pickle", "rb"))
videoOutput = ('/content/drive/MyDrive/output/demo_outPut.mp4')
mean = np.array([123.68, 176.779, 103.939], dtype= "float32")
Queue = deque(maxlen=128)
```

```python
capture_video = cv2.VideoCapture("/content/drive/MyDrive/output/sw.mp4")
writer = None
(width, height) = (None, None)

while True:
    (taken, frame) = capture_video.read()
    if not taken:
        break
    if width is None or height is None:
        (width, height) = frame.shape[:2]

    output = frame.copy()
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    frame = cv2.resize(frame, (244, 224)).astype("float32")
    frame -= mean
    # print(frame.shape)

    preds = model.predict(np.expand_dims(frame, axis=0))[0]
    Queue.append(preds)
    results = np.array(Queue).mean(axis=0)
    i = np.argmax(results)
    label = lb.classes_[i]
    text = 'PLAYING: {}'.format(label)
    cv2.putText(output, text, (45, 60), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 0), 4)

    if writer is None:
        fourcc = cv2.VideoWriter_fourcc(*'MJPG')
        writer = cv2.VideoWriter(videoOutput, fourcc, 30, (width, height), True)  # Corrected file path
    writer.write(output)
    # cv2.imshow(output)
    key = cv2.waitKey(1) & 0xFF

    if key == ord("q"):
        break

    writer.write(output)

print("Finalizing>>>>")
writer.release()
capture_video.release()
cv2.destroyAllWindows()
```
